```python
import os
from pydub import AudioSegment

# Input and Output folders
input_folder = r"D:/FYP\Datasets\Drone Datasets\Acoustic Print Dataset"
output_folder = r"D:/FYP\Datasets\Drone Datasets\Acoustic Print Dataset_Segmented"

# Create output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# Loop through each file in the input folder
for filename in os.listdir(input_folder):
    if filename.endswith(".wav"):
        file_path = os.path.join(input_folder, filename)
        audio = AudioSegment.from_wav(file_path)

        duration_ms = len(audio)  # duration in milliseconds

        if duration_ms <= 1000:
            # Save file as it is
            output_path = os.path.join(output_folder, filename)
            audio.export(output_path, format="wav")
        else:
            # Segment into 1 sec (1000 ms) chunks
            for i in range(0, duration_ms, 1000):
                segment = audio[i:i+1000]  # take 1 second
                segment_name = f"{os.path.splitext(filename)[0]}_part{i//1000 + 1}.wav"
                output_path = os.path.join(output_folder, segment_name)
                segment.export(output_path, format="wav")

print("✅ Segmentation complete. Files saved in:", output_folder)
```

```python
import os
import librosa
import soundfile as sf  # to save wavs
# Input and Output folders
input_folder = r"D:\FYP\Datasets\Drone Datasets\Salford Uni Drone Audio"
output_folder = r"D:/FYP\Datasets\Drone Datasets\Salford Uni Drone Audio_Segmented"

os.makedirs(output_folder, exist_ok=True)

SAMPLE_RATE = 16000
SEGMENT_SEC = 1.0

for fname in os.listdir(input_folder):
    if not fname.endswith(".wav"):
        continue

    path = os.path.join(input_folder, fname)
    signal, sr = librosa.load(path, sr=SAMPLE_RATE)

    length = len(signal) / sr
    if length <= SEGMENT_SEC:
        sf.write(os.path.join(output_folder, fname), signal, sr)
    else:
        n_samples = int(SEGMENT_SEC * sr)
        for i in range(0, len(signal), n_samples):
            chunk = signal[i:i+n_samples]
            if len(chunk) == 0:
                continue
            out_name = f"{os.path.splitext(fname)[0]}_part{i//n_samples+1}.wav"
            sf.write(os.path.join(output_folder, out_name), chunk, sr)

print("✅ Segmentation done.")
```

```python
import os
import re
import numpy as np
import librosa

# ========= PATHS (your provided paths) =========
DATA_FOLDER   = r"D:\FYP\Datasets\Drone Datasets\Salford Uni Drone Audio_Segmented"
CALIB_FOLDER  = r"D:\FYP\Datasets\Drone Datasets\Calibration files\Salford Uni Cal File"
OUTPUT_FOLDER = r"D:\FYP\Datasets\Drone Datasets\Salford_Uni_MFCC"
os.makedirs(OUTPUT_FOLDER, exist_ok=True)

# ========= SETTINGS =========
SAMPLE_RATE     = 16000
N_MFCC          = 13
MAX_FRAMES      = 40
MIN_DURATION_S  = 0.400  # skip very short files
PEAK_LIMIT      = 0.999  # cap gain to avoid clipping
CALIB_REQUIRED  = True   # if a mic has no calibration -> skip (True) or proceed uncalibrated (False)

# ========= FILENAME PARSING =========
# Example data names: 46212195_Ed_..._M5_part14.wav  OR  Ed 3p ... M1 part6.wav
MIC_TOKEN = re.compile(r'(?:(?<=^)|(?<=[_\s-]))(M[1-9])(?:(?=$)|(?=[_\s-]))', re.IGNORECASE)

def find_mic_token(name: str):
    """
    Return 'M1'..'M9' from the base filename if present (case-insensitive),
    where the token is separated by underscore/space/hyphen or at start/end.
    """
    base = os.path.splitext(name)[0]
    m = MIC_TOKEN.search(base)
    return m.group(1).upper() if m else None

# ========= AUDIO HELPERS =========
def load_wav_mono(path: str, sr: int = SAMPLE_RATE) -> np.ndarray:
    """
    Load mono at target SR using librosa. Returns float32 in [-1, 1].
    """
    y, _ = librosa.load(path, sr=sr, mono=True)
    return y.astype(np.float32)

def rms_linear(x: np.ndarray) -> float:
    if x.size == 0:
        return 0.0
    return float(np.sqrt(np.mean(x**2)))

def calibrate_to_target_rms(x: np.ndarray, target_rms: float, peak_limit: float = PEAK_LIMIT):
    """
    Scale x so its RMS matches target_rms, but constrain peak to <= peak_limit.
    Returns (scaled_signal, applied_gain).
    """
    cur_rms = rms_linear(x)
    if cur_rms == 0.0 or target_rms <= 0.0:
        return x, 0.0  # silence or invalid target
    desired_gain = target_rms / cur_rms

    peak = np.max(np.abs(x)) if x.size else 0.0
    if peak > 0.0:
        max_gain_no_clip = peak_limit / peak
        applied_gain = min(desired_gain, max_gain_no_clip)
    else:
        applied_gain = desired_gain

    return x * applied_gain, float(applied_gain)

def extract_mfcc_13x40x1(y: np.ndarray, sr: int = SAMPLE_RATE) -> np.ndarray:
    """Compute MFCC and pad/trim to (13, 40, 1)."""
    mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=N_MFCC)  # (13, T)
    T = mfcc.shape[1]
    if T < MAX_FRAMES:
        mfcc = np.pad(mfcc, ((0, 0), (0, MAX_FRAMES - T)), mode="constant")
    else:
        mfcc = mfcc[:, :MAX_FRAMES]
    return np.expand_dims(mfcc, axis=-1)  # (13, 40, 1)

# ========= CALIBRATION LOADING =========
# Calibration files look like: Calib_Scotland_ch1.wav ... Calib_Scotland_ch9.wav
# We'll map ch1->M1, ch2->M2, ..., ch9->M9.
CALIB_REGEX = re.compile(r'calib.*_ch([1-9])$', re.IGNORECASE)  # accepts names w/o extension too

def scan_calibration_rms(calib_dir: str) -> dict:
    """
    Build a dict mapping 'M1'..'M9' -> calibration RMS (linear) from CALIB_FOLDER.
    Prefers files with .wav, but will work if extension is .WAV or mixed case.
    """
    mic_rms = {}
    for fname in os.listdir(calib_dir):
        name_noext, ext = os.path.splitext(fname)
        if ext.lower() != ".wav":
            # ignore non-wav files
            continue
        m = CALIB_REGEX.search(name_noext)
        if not m:
            continue
        ch_num = int(m.group(1))
        mic_key = f"M{ch_num}"
        path = os.path.join(calib_dir, fname)
        y = load_wav_mono(path, sr=SAMPLE_RATE)
        cal_rms = rms_linear(y)
        mic_rms[mic_key] = cal_rms
        print(f"[CAL] {mic_key}: '{fname}' -> RMS={cal_rms:.6f}")

    # Warn if any mic missing calibration
    for i in range(1, 10):
        key = f"M{i}"
        if key not in mic_rms:
            print(f"[WARN] No calibration file found for {key}")
    return mic_rms

# ========= MAIN =========
def process_dataset(data_dir: str = DATA_FOLDER, calib_dir: str = CALIB_FOLDER, out_dir: str = OUTPUT_FOLDER):
    mic_target_rms = scan_calibration_rms(calib_dir)

    skipped = []
    for fname in os.listdir(data_dir):
        if not fname.lower().endswith(".wav"):
            continue
        in_path = os.path.join(data_dir, fname)

        mic = find_mic_token(fname)
        if mic is None:
            msg = "no_mic_token_in_name"
            print(f"[SKIP] {fname} ({msg})")
            skipped.append(f"{fname} ({msg})")
            continue

        if mic not in mic_target_rms:
            msg = f"no_calibration_for_{mic}"
            if CALIB_REQUIRED:
                print(f"[SKIP] {fname} ({msg})")
                skipped.append(f"{fname} ({msg})")
                continue
            else:
                print(f"[WARN] {fname}: {msg}. Proceeding uncalibrated.")

        try:
            # 1) Load mono at target SR
            y = load_wav_mono(in_path, sr=SAMPLE_RATE)

            # 2) Duration check
            duration_s = len(y) / SAMPLE_RATE
            if duration_s < MIN_DURATION_S:
                msg = f"too_short_{duration_s*1000:.1f}ms"
                print(f"[SKIP] {fname} ({msg})")
                skipped.append(f"{fname} ({msg})")
                continue

            # 3) Apply mic-specific relative calibration
            if mic in mic_target_rms:
                target_rms = mic_target_rms[mic]
                y, gain = calibrate_to_target_rms(y, target_rms, peak_limit=PEAK_LIMIT)
            else:
                gain = 0.0

            # 4) MFCC (13, 40, 1)
            feat = extract_mfcc_13x40x1(y, sr=SAMPLE_RATE)

            # 5) Save
            out_name = os.path.splitext(fname)[0] + "_mfcc.npy"
            out_path = os.path.join(out_dir, out_name)
            np.save(out_path, feat)

            print(f"[OK] {fname} ({mic}) -> {out_name}  {feat.shape}  gain={gain:.3f}")

        except Exception as e:
            print(f"[ERR] {fname}: {e}")
            skipped.append(f"{fname} (error: {e})")

    # Summary + log skipped
    print(f"\n✅ Done. MFCCs saved in: {out_dir}")
    if skipped:
        print("\n⚠️ Skipped files:")
        for s in skipped:
            print(" -", s)
        with open(os.path.join(out_dir, 'skipped_files.txt'), 'w', encoding='utf-8') as f:
            for s in skipped:
                f.write(s + "\n")

if __name__ == "__main__":
    process_dataset()
```

```python
import os
import librosa
import soundfile as sf  # to save wavs
# Input and Output folders
input_folder = r"D:/FYP\Datasets/Noise\Acoustic Print GR Noise Audio"
output_folder = r"D:/FYP\Datasets/Noise\Acoustic Print GR Noise Audio_Segmented"

os.makedirs(output_folder, exist_ok=True)

SAMPLE_RATE = 16000
SEGMENT_SEC = 1.0

for fname in os.listdir(input_folder):
    if not fname.endswith(".wav"):
        continue

    path = os.path.join(input_folder, fname)
    signal, sr = librosa.load(path, sr=SAMPLE_RATE)

    length = len(signal) / sr
    if length <= SEGMENT_SEC:
        sf.write(os.path.join(output_folder, fname), signal, sr)
    else:
        n_samples = int(SEGMENT_SEC * sr)
        for i in range(0, len(signal), n_samples):
            chunk = signal[i:i+n_samples]
            if len(chunk) == 0:
                continue
            out_name = f"{os.path.splitext(fname)[0]}_part{i//n_samples+1}.wav"
            sf.write(os.path.join(output_folder, out_name), chunk, sr)

print("✅ Segmentation done.")
```
